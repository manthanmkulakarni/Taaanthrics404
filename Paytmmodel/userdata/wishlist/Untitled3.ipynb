```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import random
from sklearn.naive_bayes import GaussianNB
```

```python
iteams=['LGTV','SamsungRefigerator','UshaIronBox','MorphyToaster','NikeShoes','Radowatch','RaybanGlasses','Garniearcream','MacBook','OnePlus6T','IphoneXs','Dell','WheelSoap','Turdal','Ketchup','GaramMasala']
iteams=np.array(iteams)
```

```python
clicks=[]
timespent=[]
noofvisits=[]
data=[]
for i in range(iteams.shape[0]):
    a=random.randrange(25,99)
    b=random.randrange(0,1000)
    c=random.randrange(1,12)
    data.append([a,b,c])
    clicks.append(a)
    timespent.append(b)
    noofvisits.append(c)
data=np.array(data)
```

```python
model=GaussianNB()
model.fit(data,iteams)
```

```python
model.predict(data)
```

```python
daincsv={"ProductName":iteams,"Click":clicks,"TimeSpent":timespent," Nob ofvisits ":noofvisits}
```

```python
datincsv=pd.DataFrame(daincsv)
```
