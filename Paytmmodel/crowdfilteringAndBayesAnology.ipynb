```python
def sort(array):
    array=np.array(array)
    for i in range(array.shape[0]):
        for j in range(array.shape[0]-i-1):
            if(array[j][1]>array[j+1][1]):
                temp=np.array(array[j])
                array[j]=array[j+1]
                array[j+1]=temp
    return array
```
