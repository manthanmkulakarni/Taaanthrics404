```python
date=[]
age=[]
for i in range(500):
    date.append(data4['signup_time'][i][:10])
    age.append(data4['age'][i])
```

```python
time=[]
pname=['ElectronicsAppliances','Fashion','ComputerAndMobile','Grossaries']
product=[]
for i in range(500):
    time.append(data4['signup_time'][i][11:])
    product.append(pname[random.randrange(1,25,3)%4])
    date[i]=date[i][0:10]
```

```python
time
```

```python
iteams={'ElectronicsAppliances':['LGTV','SamsungRefigerator','UshaIronBox','MorphyToaster'],'Fashion':['NikeShoes','Radowatch','RaybanGlasses','Garniearcream'],'ComputerAndMobile':['MacBook','OnePlus6T','IphoneXs','Dell'],'Grossaries':['WheelSoap','Turdal','Ketchup','GaramMasala']}
```

```python
productname=[]
for i in range(500):
    if(product[i]=='ElectronicsAppliances'):
        productname.append(iteams['ElectronicsAppliances'][random.randrange(1,10,1)%4])
    elif(product[i]=='Fashion'):
        productname.append(iteams['Fashion'][random.randrange(1,10,1)%4])
    elif(product[i]=='ComputerAndMobile'):
        productname.append(iteams['ComputerAndMobile'][random.randrange(1,10,1)%4])
    else:
        productname.append(iteams['Grossaries'][random.randrange(1,10,1)%4])
```

```python
color=[]
for i in range(500):
    color.append(data1['Avatar'][i])
sex=[]
for i in range(500):
    sex.append(data4['sex'][i])
```

```python
regionname=['Banglore','Hassan','Ooty','Manglore','Mysore']
region=[]
for i in range(500):
    region.append(regionname[random.randrange(1,15,2)%5])
```

```python
dataset={'Catagory':product,'Iteam':productname,'Gender':sex,'Age':age,'Time':time,'Color':color,'Date':date,'Region':region}
```

```python
dataset=pd.DataFrame(dataset)
```
