```python
dataforagebasedtime=np.array(dataforagebasedtime)
```

```python
temp=[]
genderindex={'M':1,'F':2}
for i in range(dataforagebasedtime.shape[0]):
    temp.append([dataforagebasedtime[i][1],genderindex[dataforagebasedtime[i][6]]])
temp=np.array(temp)
```

```python
temp
```

```python
timeslot=[]
for m in range(temp.shape[0]):
    if(temp[m][0]<25):
        k=17*60
    elif(temp[m][0]<35 and temp[m][1]==1):
        k=20*60
    elif(temp[m][0]<35 and temp[m][1]==2):
        k=15*60
    elif(temp[m][0]<45 and temp[m][1]==1):
        k=21*60
    elif(temp[m][0]<45 and temp[m][1]==2):
        k=18*60
    else:
        k=19*60
    
    timeslot.append(k)
timeslot=np.array(timeslot)
for m in range(timeslot.shape[0]):
    timeslot[m]=timeslot[m]+random.randrange(7,35)
for m in range(timeslot.shape[0]):
    timeslot[m]=timeslot[m]+random.randrange(1,15)
for m in range(timeslot.shape[0]):
    timeslot[m]=timeslot[m]+random.randrange(3,5)
for m in range(timeslot.shape[0]):
    timeslot[m]=timeslot[m]+random.randrange(1,3)
```

```python
timeslot
```

```python
agebasedtimedataframe={'Age':temp.T[0],'Gender':temp.T[1],'ActiveTime':timeslot}
agebasedtimedataframe=pd.DataFrame(agebasedtimedataframe)
agebasedtimedataframe.to_csv('AgeandGenderBaserBasedUagetime.csv')
```
